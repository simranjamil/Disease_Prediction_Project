```python
import pandas as pd
```

```python
# Load dataset (CSV)
df = pd.read_csv("Disease_data.csv")
```

```python
df = df.map(lambda x: x.strip() if isinstance(x, str) else x)
```

```python
df.head()
```

```python
del df['Unnamed: 7']
```

```python
df.head()
```

```python
# -----------------------------
# 2. Standardize text
# -----------------------------
df = df.replace({
    'yes': 'Yes', 'no': 'No',
    'male': 'Male', 'female': 'Female',
    'low': 'Low', 'normal': 'Normal', 'high': 'High'
})
```

```python
# -----------------------------
# 3. Binary columns
# -----------------------------
binary_cols = ['Fever', 'Cough', 'Fatigue', 'Difficulty Breathing']

for col in binary_cols:
    df[col] = df[col].map({'Yes': 1, 'No': 0})
```

```python
df.head()
```

```python
# -----------------------------
# 4. Gender
# -----------------------------
df['Gender'] = df['Gender'].map({'Female': 0, 'Male': 1})
```

```python
# -----------------------------
# 5. Blood Pressure
# -----------------------------
df['Blood Pressure'] = df['Blood Pressure'].map({'Low': 0, 'Normal': 1, 'High': 2})
```

```python
# -----------------------------
# 6. Cholesterol
# -----------------------------
df['Cholesterol Level'] = df['Cholesterol Level'].map({'Low': 0, 'Normal': 1, 'High': 2})
```

```python
df.head()
```

```python
df.head(10)
```

```python
# -----------------------------
# 7. Check missing values
# -----------------------------
print("Missing after mapping:\n", df.isnull().sum())
```

```python
# -----------------------------
# 8. Drop missing
# -----------------------------
df.dropna(inplace=True)
```

```python
# -----------------------------
# 9. Save cleaned data
# -----------------------------
df.to_csv("cleaned_Dieases_data.csv", index=False)
```
